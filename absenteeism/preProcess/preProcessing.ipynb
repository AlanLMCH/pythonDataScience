### Dividing the reasons into groups to do a better classification 

We need to divide the reasons into 4 groups each one being a different dataframe

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

### Concatenate Column Values

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4],axis = 1)
df.head()
```

```python
column_names = df.columns.values
```

```python
column_names
```

```python
for name in range(len(column_names)):
    if column_names[name] == 0:
        column_names[name] = 'Reason_1'
    if column_names[name] == 1:
        column_names[name] = 'Reason_2'
    if column_names[name] == 2:
        column_names[name] = 'Reason_3'
    if column_names[name] == 3:
        column_names[name] = 'Reason_4'
```

### Date

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date'].head()
```

```python
df_reason_mod.head()
```

### Extract the month value

```python
list_months = []
list_months
```

```python
total_rows = df_reason_mod.shape[0]
```

```python
for i in range(total_rows):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod.head(20)
```

### Extract day of the week

We will change monday to 0, tuesday to 1 and so on

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_date_mod = df_reason_mod.copy()
```

```python
df_reason_date_mod.head()
```
