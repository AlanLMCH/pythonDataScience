```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.cluster import MiniBatchKMeans

```

```python
datasetCandy = pd.read_csv('../data/rawData/candy.csv')

datasetCandy.head()
```

```python
datasetCandy.describe()
```

```python
x = datasetCandy.drop('competitorname', axis=1)
```

```python
kmeans = MiniBatchKMeans(n_clusters = 4, batch_size = 8).fit(x)

print('Total de centroides: ', len(kmeans.cluster_centers_))
```

```python
print(kmeans.predict(x))
```

```python
datasetCandy['group'] = kmeans.predict(x)
sns.pairplot(datasetCandy, hue='group')
```

```python
sns.scatterplot(data=datasetCandy, x="sugarpercent", y="winpercent", hue="group",palette="deep")
```
